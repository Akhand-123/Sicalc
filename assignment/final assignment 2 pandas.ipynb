### 3. Low Level Data Understanding
- Find count of unique values in loacation column.
- Find which continent has maximum frequency using values counts.
- Find maximum and mean value in 'total_cases'.
- Find 25%,50%,& 75% quartile value in 'total_deaths'.
- Find which continent has maximum. 'human_development_index'.
- Find which continent has minimum 'gdp_per_capita'.

```python
# Find count of unique values in loacation column.
df.location.count()
```

```python
# Find which continent has maximum frequency using values counts.
df.continent.value_counts()
```

Europe has the maximum frequency of values counts.

```python
# Find maximum and mean value in 'total_cases'.
df.total_cases.agg(['max','min'])
```

```python
# Find 25%,50%,& 75% quartile value in 'total_deaths'.
df.total_deaths.quantile([0.25,0.50,0.75,1])
```

```python
# Find which continent has maximum. 'human_development_index'.
df.groupby('continent',as_index=False)['human_development_index'].sum().sort_values('human_development_index',ascending=False)
```

Europe has the maximum human_development_index.

```python
# Find which continent has minimum 'gdp_per_capita'.
df.groupby('continent',as_index=False)['gdp_per_capita'].sum().sort_values('gdp_per_capita',ascending=False)
```

Oceania has the minimum gdp_per_capita.
